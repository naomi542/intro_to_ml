```python
# import standard libraries

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC # "Support vector classifier"
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# load train data
train_data = pd.read_csv('../input/classification-with-non-deep-classifiers/train.csv')
# different classes
num_labels= train_data.label.unique()
print(num_labels)
train_data.head()

#convert pandas DataFrame to numpy array if it needed!
#train_data_numpy = train_data.to_numpy()
#print(train_data_numpy.shape)
```

```python
# Data preparation, PCA and Splitting
X = train_data.iloc[:,1:].values # 0-783 features
Y = train_data.iloc[:,0].values   # class labels
#print(X)

#PCA
pca= PCA(0.85) #choose number of components such that 85% of variance is explained
pca.fit(X)
X_reduced= pca.transform(X) #use this in machine learning model
print(pca.n_components_) 
#split data into training set and validation set to 'test' the accuracy of our model before it is used on actual test set
# split the training data into 80% train set and 20% validation set
#when submitting final model train on all the data
X_train_set, X_validation_set, Y_train_set, Y_validation_set = train_test_split(X_reduced, Y, test_size=0.20)
```

```python
# Create SVM Classifier 
model = SVC(kernel='rbf', C=200)
#model.fit(X_train_set, Y_train_set) #when only using 80% of training data 
model.fit(X_reduced, Y) #when using 100% of training data
```

```python
#predict validation data- comment out for actual submission
Y_predictions = model.predict(X_validation_set)
accuracy_score(Y_validation_set, Y_predictions)
```
